```python
from audiocraft.models import MusicGen
from audiocraft.utils.notebook import display_audio
import torch
```

```python
model = MusicGen.get_pretrained('facebook/musicgen-melody', device='cuda')
model.set_generation_params(duration=15)
```

```python
musics = model.generate(
  [
      'earthy tones, environmentally conscious, ukulele-infused, harmonic, breezy, easygoing, organic instrumentation, gentle grooves',
      'Create a musical backdrop that embodies earthy tones and an environmentally conscious spirit, featuring a gentle ukulele as the central melodic voice. Weave in harmonic layers that feel breezy and easygoing, using organic instrumentation that breathes with natural, soft rhythms. Craft subtle, gentle grooves that evoke the feeling of a warm, sunlit afternoon in a lush, green landscape - calm yet subtly energetic, like a light breeze rustling through leaves. The music should feel intimate and unpretentious, with a sense of peaceful connection to the natural world.'
  ],
  progress = True)
display_audio(musics, 32000)
```
